### 2.1 calculating the age distribution among the unique customers

```python
def get_unique_rows(column_name, final_df=final_df):
    """
    The function looks for unique rows from the dataframe by means of person id
    """
    unique_id = final_df['id_x'].unique()
    value = []
    for idx in unique_id:
        value.append(final_df[final_df['id_x']==idx][column_name].values[0])
        
    return pd.DataFrame({column_name: np.array(value)})
```

```python
def get_age_range(v):
    """
    Create age range
    """
    
    ranges = [(0,20),(21,30), (31,40), (41,50),(51,60), (61,70),(71,80), (81,90),(91, 100), (101, 110),(111, 120)]
    for elements in ranges:
        if elements[0] <= v and v <= elements[1]:
            return f'{elements[0]}-{elements[1]}'
    return np.nan
```

# 3 Machine Learning Implementations

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import minmax_scale, LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
```

### 3.1 Building the benchmark model, KNN model

```python
# creating new names for the offers
keep_count = {'bogo': 0,'discount': 0}
new_offer_names = []
for offer_type in portfolio[portfolio.offer_type!='informational'].set_index('id').offer_type.values:
    new_offer_names.append(f'{offer_type}_{keep_count[offer_type]}')
    keep_count[offer_type]+=1
    
print(new_offer_names)
```

```python
new_portfolio = portfolio[portfolio.offer_type!='informational']
```

```python
new_portfolio['new_offer_names']=new_offer_names
```

```python
df2=pd.merge(final_df[(final_df.event!='transaction') | (final_df.offer_type!='informational')], new_portfolio[['id','new_offer_names']],\
         left_on='new_value', right_on='id')
```

```python
le_gender = LabelEncoder()
le_gender.fit(df2['gender'])
df2['gender_n']=le_gender.transform(df2['gender'])
```

```python
def splitChannels(df):
    channels = {'web':[],
               'email':[],
               'mobile':[],
               'social':[]}
    
    for chnl_list in df['channels']:
        for key_chnl in channels.keys():
            if key_chnl in chnl_list:
                channels[key_chnl].append(1)
            else:
                channels[key_chnl].append(0)
    return pd.DataFrame(channels)
channels_df=splitChannels(df2)
```

```python
df2['web']=channels_df['web']
df2['email']=channels_df['email']
df2['mobile']=channels_df['mobile']
df2['social']=channels_df['social']
```

```python
def getBenchmarkModelFeatures(df, test_percentage=0.4, seed=100):
    """This functions takes in dataframe as input and return training and testing data
    
    Paremeters
    ----------
        df: `pandas DataFrame`
        
        test_percentage: float
            The percentage of the data that will be used for training. ranges 0-1
            
        seed: int
            The value to use to always get a specific randomness
    Returns
    -------
    """
    #making dummy columns |event_offer completed|event_offer received|event_offer viewed|
    new_df_with_dummies = pd.get_dummies(df, columns=['event']).dropna()
    
    #The lines below are trying to get rows that show where an offer has been completed of not
    mean_df = new_df_with_dummies.groupby(['id_x', 'new_offer_names']).mean().drop(
              columns=['time','event_offer completed',
                       'event_offer received',
                       'event_offer viewed'])
    
    offers_df = new_df_with_dummies.groupby(['id_x','new_offer_names']).sum()[
                ['event_offer completed',
                 'event_offer received',
                 'event_offer viewed']
                ]
    
    formatted_df = mean_df.merge(offers_df, left_on=['id_x','new_offer_names'], 
                                 right_on=['id_x','new_offer_names'])\
                                 .reset_index(level=['id_x','new_offer_names'])
#     formatted_df = formatted_df[formatted_df['new_offer_names']=='bogo_0']
    #preparing the data for training
    X_features = formatted_df[['age','income','reward','difficulty',
                               'duration','gender_n','web',
                               'email','mobile','social']].values
    y_labels = formatted_df['event_offer completed'].apply(lambda x: 1 if x>=1 else 0).values # converting to 
    
    #Using normalization to scale the features
#     X_features = minmax_scale(X_features)

    #Using standard Scaler
    scaler = StandardScaler()
    scaler.fit(X_features)
    X_features = scaler.transform(X_features)
    
    #spliting the data for training
    X_train, X_test, y_train, y_test = train_test_split(X_features, 
                                                        y_labels, 
                                                        test_size=test_percentage, 
                                                        random_state=seed)

    
    return X_train, X_test, y_train, y_test
    
    
    
X_train, X_test, y_train, y_test= getBenchmarkModelFeatures(df2)
```

### 3.2 KNN model

```python
# calculating the accuracy of models with different values of k
k_neighbors_range = 100 #starting from 1
mean_acc = np.zeros(k_neighbors_range)
models = []
for i in range(1,k_neighbors_range+1):
    #Train Model and Predict  
    knn = KNeighborsClassifier(n_neighbors = i).fit(X_train,y_train)
    models.append(knn)
    yhat= knn.predict(X_test)
    mean_acc[i-1] = accuracy_score(y_test, yhat)
```

```python
loc = np.arange(1,k_neighbors_range+1,step=1.0)
plt.figure(figsize = (20, 12))
plt.plot(range(1,k_neighbors_range+1), mean_acc)
plt.grid()
plt.xticks(loc, rotation=90)
plt.title('KNN(Benchmark model)')
plt.xlabel('Number of Neighbors ')
plt.ylabel('Accuracy')
plt.show()
```

```python
print(f'The best knn model has an accuracy of {round(mean_acc.max()*100,2)}% and use {mean_acc.argmax()+1} neighbors')
```

```python
best_knn_model = models[mean_acc.argmax()]
predictions = best_knn_model.predict(X_test)
print(classification_report(y_test, predictions))
```

- #### The benchmark model above is trained by changing the number of neighbors hyperparameter and choosing the best one
- #### The above benchmark model is trained to find out if a customer would complete or not complete an offer given to them
- #### The model can then be used to figure out which offer to give to a customer that they would complete
- #### The model has an accuracy of 69% and 71% precision towards customers completing an offer
- #### The above model has very low recall for customer who would not complete the given offer, the model could simply be guessing that majority of the customer will complete an offer

### 3.3 Decision Tree Model

```python
def chooseSearchGrid(model,kind, parameters,X_train,y_train):
    """
    This function tunes the hyperparameters of models using grid search or random search
    
    
    Parameters
    ----------
        model: 
            This is the model that must be tuned
        kind: str
            The values helps decide to choose a grid search or random search
        parameters: dict
            Contains the hyperparameters that must be tuned
        X_train: numpy.ndarry
            The training features
        y_train: numpy.ndarray
            The target features
            
    Returns
    -------
        The best possible estimator from the given hyperparameters."""
    
    if kind=='grid':
        search_obj = GridSearchCV(model, param_grid=parameters)
        search_obj.fit(X_train,y_train)
        best_estimator = search_obj.best_estimator_
        print(best_estimator)
        
    else:
        search_obj = RandomizedSearchCV(model, param_distributions=parameters)
        search_obj.fit(X_train,y_train)
        best_estimator = search_obj.best_estimator_
        print(best_estimator)
        
    return best_estimator

parameters = { 
    'max_depth':[1,2,3,4,5,6],
    'min_samples_leaf':[1,2,3,4,5,6],
    'min_samples_split': [2,3,4,5,6],
    'criterion':['gini','entropy']
}
```

```python
d_tree = DecisionTreeClassifier(random_state=10)
```

#### Choosing the best decision tree from a grid search

```python
best_estimator = chooseSearchGrid(d_tree, 'grid', parameters, X_train, y_train)
```

```python
prediction=best_estimator.predict(X_test)
```

```python
print(classification_report(y_test, prediction))
```

#### Choosing the best decision tree from a random search

```python
best_estimator = chooseSearchGrid(d_tree, 'random', parameters, X_train, y_train)
```

```python
prediction=best_estimator.predict(X_test)
```

```python
print(classification_report(y_test, prediction))
```

### 3.4 RandomForest Model

```python
forest = RandomForestClassifier(random_state=10)
```

#### Choosing the best Random forest classifier from a grid search

```python
best_estimator = chooseSearchGrid(forest, 'grid', parameters, X_train, y_train)
```

```python
prediction=best_estimator.predict(X_test)
```

```python
print(classification_report(y_test, prediction))
```

#### Choosing the best Random forest classifier from a random search

```python
best_estimator = chooseSearchGrid(forest, 'random', parameters, X_train, y_train)
```

```python
prediction=best_estimator.predict(X_test)
print(classification_report(y_test, prediction))
```
